#1. Charge estimates by whole tracker and inner tracker

```python
args = {
    'bins': np.linspace(0,5,100),
    'histtype':'step', 'lw':2, 'ax': gca()
}
for n in ['Q_all', 'InnerQ_all']:
    data[n][data[n]>0].hist(label=n, **args)
yscale('log')
legend()
```

#2.Charge estimation by individual layers

```python
args = { 'bins': np.linspace(0,5,100), 'histtype':'step', 'lw':2, 'ax': gca() }
colors = {
    '0':'k', '1':'r', '2':'b', '3':'b', '4':'b', '5':'b', '6':'b', '7':'b', '8':'#aaaaaa' }

for n in [c for c in data.columns if c.startswith('LayerJQ')]:
    data[n][data[n]>0].hist(label=n, color=colors[n[-1]], **args)
yscale('log')
legend()
```
