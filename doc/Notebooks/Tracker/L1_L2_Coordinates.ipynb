```python
figsize(10,5)
for n,name in  enumerate(['L1','L2']): 
    cols = [c for c in data.columns if c.startswith(name)]
    subplot(1,2,n+1)
    args = {
        'kind':'scatter', 'x': name + "_Hit_X", 'y': name + "_Hit_Y",
        'marker':'.', 'lw':0, 's':1, 'ax':gca()
    }
    data[(data[cols] != 0).any(axis=1)].plot(**args)
    xlim(-80,80); ylim(-60,60);
tight_layout()
```
