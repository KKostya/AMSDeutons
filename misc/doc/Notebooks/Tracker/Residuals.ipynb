#1 Residuals

```python
Here are the residuals for the inner tracker fit:
```

```python
kwds = { "bins":np.linspace(-0.1,0.1,100), "histtype":"stepfilled", "lw":1, }
figsize(15,10)
N = 9
for i in range(N):
    subplot(N,2,2*i+1, yticklabels=[]); title("X side, Layer "+str(i+1))
    data["ResidualX_"+str(i)][data['R']!=0].hist(**kwds); 
    setp(gca().get_xticklabels(), visible=(i==N-1));xlim(-0.1,0.1)
    subplot(N,2,2*i+2, yticklabels=[]); title("Y side, Layer "+str(i+1))
    data["ResidualY_"+str(i)][data['R']!=0].hist(**kwds)
    setp(gca().get_xticklabels(), visible=(i==N-1));xlim(-0.1,0.1)
tight_layout()
```
